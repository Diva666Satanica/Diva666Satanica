```python
def PathToGenome(kmerList):

  seq = kmerList[0]
  n = len(kmerList)
  for i in range(n-1):
    seq = seq + kmerList[i+1][-1]
  return seq

PathKmers = '''ACCGA
CCGAA
CGAAG
GAAGC
AAGCT'''
kmers = PathKmers.split('\n')
print(kmers)
out = PathToGenome(kmers)
print(out)
```

Week1 : 1.3.2 - Creating the adjacency matrix for representing a unidirectional map

```python
def Overlap(patterns):
  adjacency = {}
  n = len(patterns)
  for kmer in patterns:
    patN = len(kmer)
    adjacency[kmer] = []
    for i in range(n):
      if kmer != patterns[i]:
        if kmer[1:patN] == patterns[i][0:patN-1]:
          adjacency[kmer].append(patterns[i])
  return adjacency

patterns = '''ATGCG
GCATG
CATGC
AGGCA
GGCAT
GGCAC'''
pat = patterns.split('\n')
print(pat)
out = Overlap(pat)
print(out)
finout =''
for key, value in out.items():
  if len(value) > 0:
    finout = finout + key + ' -> ' + ','.join(value) + '\n'
print(finout)
```

```python
def DeBruijn(Text, k):
  adjacency_dict = {}
  adjacency_list = ''
  for i in range(len(Text)-k+1):
    entry = Text[i:i+k]
    try:
      adjacency_dict[entry[0:k-1]].append(entry[1:k])
    except KeyError:
      adjacency_dict[entry[0:k-1]] = [entry[1:k]]

  for key, value in adjacency_dict.items():
    if len(value) > 0:
      adjacency_list = adjacency_list + key + ' -> ' + ','.join(value) + '\n'

  return adjacency_dict, adjacency_list

k = 4
Text = "AAGATTCTCTAAGA"

out = DeBruijn(Text, k)
print(out[0])
print(out[1])
```
